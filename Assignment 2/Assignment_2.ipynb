# ***Working with another Dataset (titanic_train.csv)***

*Importing modules*

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from seaborn import load_dataset
```

```python
data = pd.read_csv("titanic_train.csv")
```

```python
data.head()
```

```python
data.tail()
```

## **Categorical Data**

1. ***Count Plot***

```python
# sns.countplot(data['Survived'])
# plt.show()
sns.countplot(x='Survived', data=data)
plt.show()

# Countplot is used to visualize the frequency of categorical data.
# 'Survived' is a categorical target variable (0 = Not Survived, 1 = Survived).
# We use x='Survived' and data=data to explicitly tell Seaborn which column to count and plot on the x-axis.
# This produces a bar graph showing the number of passengers who survived and who did not survive, similar to pandas value_counts() visualization.
```

2. ***Pie Chart***

```python
data['Sex'].value_counts().plot(kind="pie",autopct="%.2f")
plt.ylabel('Sex') # Explicitly set label
plt.show()
```

## **Numerical Data**

1. ***Histogram***

```python
plt.hist(data['Age'], bins=5)
plt.show()
```

2. ***Distplot***

```python
sns.distplot(data['Age'])
plt.show()
```

3. ***Boxplot***

```python
# Boxplot is used to visualize the five-number summary of numerical data.
# It displays the median, first quartile (Q1), third quartile (Q3), and detects outliers using the interquartile range (IQR).
# Here, the boxplot is used to analyze the distribution of passenger ages in the Titanic dataset.

sns.boxplot(y=data['Age'])
plt.title('Boxplot of Age')
plt.ylabel('Age')
plt.show()

# Age is chosen because it is a numerical continuous variable suitable for showing median, quartiles, IQR, and outliers using a boxplot.
# Boxplots are specifically used to analyze numerical data and to visualize the five-number summary (minimum, Q1, median, Q3, maximum).
# Age also contains outliers in the Titanic dataset, which makes it suitable for demonstrating the use of Interquartile Range (IQR).
# Hence, Age is an appropriate column to understand data distribution and outliers using a boxplot.
```

## **Scatter Plot**

```python
tips = sns.load_dataset("tips")

# The load_dataset() function loads a built-in dataset provided by Seaborn.
# Here, a new dataset named "tips" is loaded, which contains information
# about total bill amount and tip given in a restaurant.

sns.scatterplot(x="total_bill", y="tip", data=tips)


# Scatter plot is used to analyze the relationship between two numerical variables.
# 'total_bill' is plotted on the x-axis and 'tip' is plotted on the y-axis.
# Explicitly specifying x, y, and data is required in newer versions of Seaborn, which is why this code works correctly.

plt.show()

# sns.scatterplot(tips["total_bill"], tips["tip"])
# plt.show()

# This code uses positional arguments for x and y.
# In older versions of Seaborn, positional arguments were allowed, but in newer versions, Seaborn requires explicit axis mapping.
# Since x and y are not specified, Seaborn cannot determine which variable belongs to which axis, resulting in an error.
```

***Multivariate analysis with scatter plot***

```python
# sns.scatterplot(tips["total_bill"], tips["tip"], hue=tips["sex"])
# plt.show()

# The first (commented) code does not work because it uses positional arguments for x and y (tips["total_bill"], tips["tip"]).
# In newer versions of Seaborn, scatterplot() does not allow multiple positional arguments for x and y, especially when keyword arguments like 'hue' are used.
# This causes a TypeError because Seaborn cannot determine axis mapping.

sns.scatterplot(x="total_bill", y="tip", hue="sex", data=tips)
plt.show()

# The second code works because x, y, and hue are explicitly specified as column names and the dataset is provided using data=tips.
# This clearly tells Seaborn which columns to use for the x-axis, y-axis, and color grouping, making the code compatible with
# the latest Seaborn versions.
```

```python
# sns.scatterplot(tips["total_bill"], tips["tip"], hue=tips["sex"], style=tips["smoker"])
# plt.show()

# In newer versions of Seaborn, scatterplot() does not allow multiple positional arguments for x and y, especially when keyword arguments like 'hue' are used.

sns.scatterplot(x="total_bill", y="tip", hue="sex", style="smoker", data=tips)
plt.show()
```

3. ***Distplot***

```python
# sns.displot(data[data['Survived']==0]['Age'], hist=False, color="blue")
# sns.displot(data[data['Survived']==1]['Age'], hist=False, color="orange")
# plt.show()


sns.kdeplot(data=data[data['Survived'] == 0], x='Age', label='Not Survived')
sns.kdeplot(data=data[data['Survived'] == 1], x='Age', label='Survived')
plt.legend()
plt.show()

# KDE plot is used to visualize the distribution of a numerical variable.
# Here, Age distribution is compared for survived and not survived passengers.
# sns.kdeplot() is an axes-level function, so multiple KDE plots can be overlaid on the same graph.
# sns.displot() is a figure-level function and cannot be used for overlaying multiple distributions, which is why it was not working.
```
