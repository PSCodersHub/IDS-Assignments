````python
# Importing required libraries for numerical computation,
# data manipulation, and visualization
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
# URL of the Iris dataset from UCI Machine Learning Repository
# This dataset is used only for practice purposes
csv_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'

# Assigning column names because the dataset does not contain headers
col_names = [
    'sepal_length',
    'sepal_width',
    'petal_length',
    'petal_width',
    'species'
]

# Loading the Iris dataset into a Pandas DataFrame
# 'names' parameter is used to add column headers
iris = pd.read_csv(csv_url, names=col_names)

# Displaying the first five rows to understand dataset structure
iris.head()
````

**Class-wise Descriptive Statistics**

## **ðŸ“Š DESCRIPTIVE STATISTICS & POWER TRANSFORMATION**
Main Implementation Using Diabetes Dataset

**Objective**

*To perform descriptive statistical analysis on a regression dataset, apply Linear Regression before and after power transformation, and study the effect of Yeo-Johnson transformation on model performance.*

***Step-1: Import Required Libraries***

````python
# Importing libraries for numerical computation and data handling
import numpy as np
import pandas as pd

# Importing libraries for data visualization
import seaborn as sns
import matplotlib.pyplot as plt

# Importing statistical functions
import scipy.stats as stats

# Suppressing warnings for clean output
import warnings
warnings.filterwarnings('ignore')
````

**ðŸ”¹ Purpose:**

*These libraries help in data manipulation, statistical analysis, visualization, and model evaluation*

---

***Step-2: Import Machine Learning Packages***

````python
# Importing tools for dataset splitting and model evaluation
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score

# Importing Linear Regression model
from sklearn.linear_model import LinearRegression

# Importing performance metric
from sklearn.metrics import r2_score

# Importing Power Transformation method
from sklearn.preprocessing import PowerTransformer
````

**ðŸ”¹ Purpose:**

*To build, train, and evaluate a regression model and apply power transformation to the data.*

---

***Step-3: Load the Diabetes Dataset***

````python
# Loading the inbuilt Diabetes dataset (regression dataset)
from sklearn.datasets import load_diabetes

diabetes = load_diabetes()

# Converting dataset into Pandas DataFrame
df = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)

# Adding target variable to the DataFrame
df['Target'] = diabetes.target

# Displaying first five rows of the dataset
df.head()
````

**ðŸ”¹ Purpose:**

*To load a clean, numerical regression dataset suitable for descriptive statistics and power transformation.*

---

***Step-4: Check for Missing Values***

````python
# Checking for missing values in each column
print(df.isnull().sum())
````

**ðŸ”¹ Purpose:**

*To ensure data completeness before applying statistical analysis and machine learning models.*

---

***Step-5: Descriptive Statistical Analysis***

````python
# Generating descriptive statistics for the dataset
df.describe()
````

**ðŸ”¹ Purpose:**

*To compute statistical measures such as mean, standard deviation, minimum, maximum, and quartiles for each feature.*

---

***Step-6: Separate Independent and Dependent Variables***

````python
# Separating independent variables (features)
X = df.iloc[:, :-1]

# Separating dependent variable (target)
y = df.iloc[:, -1]
````

**ðŸ”¹ Purpose:**

*To clearly distinguish input features from the output variable for regression modeling.*

---

***Step-7: Split Dataset into Training and Testing Sets***

````python
# Splitting data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=105
)
````

**ðŸ”¹ Purpose:**

*To train the model on one portion of data and test its performance on unseen data.*

---

***Step-8: Linear Regression Before Power Transformation***

````python
# Creating and training Linear Regression model
lr = LinearRegression()
lr.fit(X_train, y_train)

# Predicting target values for test data
y_pred = lr.predict(X_test)

# Calculating R2 score before transformation
r2_before = r2_score(y_test, y_pred)

print("R2 Score before transformation:", r2_before)
````

**ðŸ”¹ Purpose:**

*To evaluate the baseline performance of Linear Regression on original data.*

---

***Step-9: Distribution and Q-Q Plots (Before Transformation)***

````python
# Plotting distributions and Q-Q plots for each feature
for col in X_train.columns:
    plt.figure(figsize=(14,4))

    plt.subplot(121)
    sns.distplot(X_train[col])
    plt.title(f"{col} - Distribution")

    plt.subplot(122)
    stats.probplot(X_train[col], dist="norm", plot=plt)
    plt.title(f"{col} - Q-Q Plot")

    plt.show()
````

**ðŸ”¹ Purpose:**

*To check normality of features and identify skewness in the data before transformation.*

---

***Step-10: Apply Yeo-Johnson Power Transformation***

````python
# Applying Yeo-Johnson power transformation
pt = PowerTransformer(method='yeo-johnson')

X_train_transformed = pt.fit_transform(X_train)
X_test_transformed = pt.transform(X_test)

# Displaying lambda values for each feature
pd.DataFrame({
    'Cols': X_train.columns,
    'Lambdas': pt.lambdas_
})
````

**ðŸ”¹ Purpose:**

*To transform data into a more normal distribution while supporting zero and negative values.*

---

***Step-11: Linear Regression After Power Transformation***

````python
# Training Linear Regression model on transformed data
lr = LinearRegression()
lr.fit(X_train_transformed, y_train)

# Predicting using transformed test data
y_pred2 = lr.predict(X_test_transformed)

# Calculating R2 score after transformation
r2_after = r2_score(y_test, y_pred2)

print("R2 Score after transformation:", r2_after)
````

**ðŸ”¹ Purpose:**

*To compare model performance before and after power transformation.*

---

***Step-12: Distribution Plots After Transformation***

````python
# Converting transformed data into DataFrame for plotting
X_train_transformed = pd.DataFrame(
    X_train_transformed, columns=X_train.columns
)

# Comparing distributions before and after transformation
for col in X_train.columns:
    plt.figure(figsize=(14,4))

    plt.subplot(121)
    sns.distplot(X_train[col])
    plt.title("Before Transformation")

    plt.subplot(122)
    sns.distplot(X_train_transformed[col])
    plt.title("After Yeo-Johnson Transformation")

    plt.show()
````

**ðŸ”¹ Purpose:**

*To visually confirm improvement in data normality after applying power transformation.*

**Result**

The Diabetes dataset was analyzed using descriptive statistical measures and Linear Regression was applied before and after power transformation.

* The **RÂ² score before power transformation** was **0.5192**, indicating moderate predictive performance of the Linear Regression model on the original data.
* After applying **Yeo-Johnson power transformation**, the **RÂ² score became 0.5049**.
* Distribution plots showed that the features became more symmetric and closer to a normal distribution after transformation.
